```python

import csv
import pandas as pd
```

```python
ten=pd.read_csv('Resources/StormEvents_details-ftp_v1.0_d2010_c20220107.csv')
eleven=pd.read_csv('Resources/StormEvents_details-ftp_v1.0_d2011_c20220107.csv')
twelve=pd.read_csv('Resources/StormEvents_details-ftp_v1.0_d2012_c20220107.csv')
thirteen=pd.read_csv('Resources/StormEvents_details-ftp_v1.0_d2013_c20220124.csv')
forteen=pd.read_csv('Resources/StormEvents_details-ftp_v1.0_d2014_c20211217.csv')
fifteen=pd.read_csv('Resources/StormEvents_details-ftp_v1.0_d2015_c20211217.csv')
sixteen=pd.read_csv('Resources/StormEvents_details-ftp_v1.0_d2016_c20211217.csv')
seventeen=pd.read_csv('Resources/StormEvents_details-ftp_v1.0_d2017_c20220124.csv')
eighteen=pd.read_csv('Resources/StormEvents_details-ftp_v1.0_d2018_c20220217.csv')
ninteen=pd.read_csv('Resources/StormEvents_details-ftp_v1.0_d2019_c20220214.csv')
twenty=pd.read_csv('Resources/StormEvents_details-ftp_v1.0_d2020_c20220217.csv')
twoone=pd.read_csv('Resources/StormEvents_details-ftp_v1.0_d2021_c20220217.csv')
```

```python
master=pd.concat([ten,eleven,twelve,thirteen,forteen,
fifteen,sixteen,seventeen,eighteen,ninteen,twenty,twoone],ignore_index=True)
```

```python
master_clean=master[['EVENT_ID', 'STATE', 
       'YEAR', 'MONTH_NAME', 'EVENT_TYPE','DAMAGE_PROPERTY', 'SOURCE',
       'FLOOD_CAUSE', 'CATEGORY','BEGIN_RANGE','BEGIN_LOCATION', 'END_RANGE','END_LOCATION','EVENT_NARRATIVE']]
```

```python
master_clean.head()
```

```python
#dropping additional unnecessary columns
master_clean=master[['EVENT_ID', 'STATE', 
       'YEAR', 'MONTH_NAME', 'EVENT_TYPE',
       'FLOOD_CAUSE', 'CATEGORY','BEGIN_RANGE','BEGIN_LOCATION', 'END_RANGE','END_LOCATION','EVENT_NARRATIVE']]
```

```python
#convert series to dataframe
df = pd.DataFrame(master_clean)
df.head()
```

```python
#check for null values
df.isnull().sum()
```

```python
#drop columns with only null values
df = df.drop(['BEGIN_RANGE', 'BEGIN_LOCATION', 'END_RANGE', 'END_LOCATION'], axis=1)
df.head()
```
